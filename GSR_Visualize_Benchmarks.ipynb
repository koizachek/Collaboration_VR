Create the benchmark file

```python
import pandas as pd

# List to store the GSR data from each CSV file
gsr_data = []

# Read in the data from each CSV file and store the GSR data in the list
for i in range(1, 25):
    df = pd.read_csv(f"./Normal/gsrdata_{i}.csv")
    gsr_data.append(df["GSR"])

# Calculate the average GSR value
average_gsr = sum(gsr_data) / len(gsr_data)

# Write the average GSR data to a new CSV file
average_df = pd.DataFrame({"GSR": average_gsr})
average_df.to_csv("averaged_gsr.csv", index=True)
```

rescale files with more entries than the benchmark file

```python
# Read in the data from the CSV file
df = pd.read_csv("averaged_gsr_group7.csv", index_col='Time')
```

```python
# Create an empty list to store the mean values
mean_values = []

# Iterate over the rows of the dataframe, taking the mean of every two rows
for i in range(0, len(df), 2):
    row1 = df.iloc[i]
    row2 = df.iloc[i+1]
    mean_value = (row1['GSR'] + row2['GSR']) / 2
    mean_values.append(mean_value)

# Create a new dataframe using the mean values
df_resampled = pd.DataFrame({'GSR_mean': mean_values})
df_resampled.to_csv('gsr_scaled.csv')
```

Visualize the benchmark 

```python
# Read in the data from the CSV file
df = pd.read_csv("averaged_gsr.csv")

# Set the figure size and background color
plt.figure(figsize=(15, 6))

# Plot the GSR data
plt.plot(df["Unnamed: 0"], df["GSR"], color='darkblue')

# Add a grid to the plot
plt.grid(color='gray', linestyle='dashed')

# Add x-axis and y-axis labels
plt.xlabel('Time (ticks)')
plt.ylabel('GSR Signal')

# Show the plot
plt.show()
```
