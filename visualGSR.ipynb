# Visualization of the Normalized GSR Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('./Normal/G4_Diana_normalized_gsrdata.csv')
df.head()
```

Linechart for an individual user:

```python
# Select the column with the GSR data
gsr_data = df['GSR']

# Calculate the rolling average of the GSR data with a window size of 50 data points
rolling_average = gsr_data.rolling(50).mean()

# Add x-axis and y-axis labels
plt.xlabel('Time (ticks)')
plt.ylabel('GSR')

# Plot the rolling average of the GSR data as a line chart
plt.plot(rolling_average)
plt.show()
```

Linechart for all users

```python
import pandas as pd
import scipy.interpolate as interpolate
import seaborn as sns
```

```python
# Initialize a list to store the data from the CSV files
df_list = []

# Loop through the CSV files
for i in range(1, 25):
    # Read the CSV file and store the data in a pandas dataframe
    df = pd.read_csv('./Normal/gsrdata_{}.csv'.format(i))
    
    # Calculate the rolling average of the GSR data
    df['gsr_rolling'] = df['GSR'].rolling(window=10).mean()
    
    df_list.append(df)
    
# Concatenate the dataframes into a single dataframe
df_concat = pd.concat(df_list)
```

```python
# Plot the line chart using seaborn
sns.lineplot(x='Time', y='gsr_rolling', hue='GSR', data=df_concat)

# Adjust the size of the plot
plt.figure(figsize=(10, 6))

# Add x-axis and y-axis labels
plt.xlabel('Time (ticks)')
plt.ylabel('GSR (uV)')

# Add a legend
#plt.legend()

plt.show()
```
